```python
from gensim.models import Word2Vec
from gensim.utils import simple_preprocess

# Step 3: Preprocess your text
text = "Phones that made effective use of any significant data connectivity were still rare outside Japan until the introduction of the Danger Hiptop in 2002, which saw moderate success among U.S. consumers as the T-Mobile Sidekick. Later, in the mid-2000s, business users in the U.S. started to adopt devices based on Microsoft's Windows Mobile, and then BlackBerry smartphones from Research In Motion. American users popularized the term CrackBerry in 2006 due to the BlackBerry's addictive nature.[31] In the U.S., the high cost of data plans and relative rarity of devices with Wi-Fi capabilities that could avoid cellular data network usage kept adoption of smartphones mainly to business professionals and early adopters. Outside the U.S. and Japan, Nokia was seeing success with its smartphones based on Symbian, originally developed by Psion for their personal organisers, and it was the most popular smartphone OS in Europe during the middle to late 2000s. Initially, Nokia's Symbian smartphones were focused on business with the Eseries,[32] similar to Windows Mobile and BlackBerry devices at the time. From 2002 onwards, Nokia started producing consumer-focused smartphones, popularized by the entertainment-focused Nseries. Until 2010, Symbian was the world's most widely used smartphone operating system.[33]"  # Replace with your actual input paragraph
tokens = simple_preprocess(text)

# Step 4: Train Word2Vec model
model = Word2Vec([tokens], vector_size=10, window=5, min_count=1, workers=4, epochs = 50)

# Step 5: Access word vectors
vector = model.wv['smartphone']
vector = model.wv['android']

# Step 6: Utilize Word2Vec features
similar_words = model.wv.most_similar('smartphone')

# Print the word vector of 'smartphone'
print("Vector representation of 'smartphone':")
print(vector)

# Print the most similar words to 'smartphone'
print("\nSimilar words to 'smartphone':")
for word, similarity in similar_words:
    print(word, similarity)
```

```python
from gensim.models import Word2Vec
from gensim.models.callbacks import CallbackAny2Vec

# Step 2: Prepare training data
sentences = [
    ['Innovation', 'reliability', 'sleekness', 'and', 'performance', ' make', 'consumers', 'prefer', 'an', 'iPhone.'],
    ['Seamless', 'user-friendly', 'interface', 'advanced', 'features', 'and', 'durability', 'attract', 'consumers', 'to', 'iPhones.'],
    ['Aesthetically', 'pleasing', 'design', 'cutting-edge', 'technology', 'impressive', 'camera', 'and', 'app', 'ecosystem', 'entice', 'consumers', 'towards', 'iPhones.'],
    ['Superior', 'user', 'experience', 'seamless', 'integration', 'long-lasting', 'battery', 'life', 'and', 'frequent', 'software', 'updates', 'contribute', 'to', 'the', 'appeal', 'of', 'iPhones', 'for', 'consumers.']

]

# Step 4: Define callback class (optional)
class CustomCallback(CallbackAny2Vec):
    def on_epoch_end(self, model):
        print("Training epoch ended")

# Step 5: Train the CBOW model
embedding_size = 100
context_window = 5
min_word_count = 1
num_workers = 4


model = Word2Vec(sentences, vector_size=embedding_size, window=context_window, sg=0, epochs=20, min_count=min_word_count, workers=num_workers, callbacks=[CustomCallback()])

# Step 6: Save and load the model
model.save("cbow_model.bin")
loaded_model = Word2Vec.load("cbow_model.bin")
```
